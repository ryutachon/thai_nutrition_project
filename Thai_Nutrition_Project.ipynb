# SQLite Testing

```
import sqlite3 as sl
```

```
con = sl.connect('thaifood.db')
#thaifood.to_sql('THAI_FOOD',con)
```

THAI_FOOD is now in SQL

```
pd.read_sql('''
            SELECT menu_en, type, energy_kcal, fat_g
            FROM THAI_FOOD
            WHERE (menu_en LIKE '%fish%' OR menu_en LIKE 'Fish%')
            AND (menu_en NOT LIKE '%curry%')
            AND (energy_kcal BETWEEN 1 AND 200 OR fat_g < 10)
            ORDER BY energy_kcal, fat_g
            ''',con)
```
